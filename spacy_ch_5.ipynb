```python
import pandas as pd

data = pd.read_csv("data/fda_data.csv").set_index('drug_id')
inds_table = data['indications_and_usage']

def prune_header(ind):
    ind = str(ind)
    if ind.startswith('INDICATIONS AND USAGE: '):
        return ind[23:]
    elif ind.startswith('1 INDICATIONS AND USAGE '):
        return ind[24:]
    elif ind.startswith('INDICATIONS AND USAGE '):
        return ind[22:]
    
inds_table = inds_table.apply(prune_header)
```

```python
import spacy
nlp = spacy.load("en_core_web_md")
```

```python
# OOV terms in first 500 docs
nlp_inds = [nlp(d) for d in inds_table[:1000] if d]
{str(t) for d in nlp_inds for t in d if t.is_oov}
```

```python
# reduce to named entities and re-process
ind_ents = [nlp(" ".join([str(e) for e in d.ents])) for d in nlp_inds]
ind_ents = [e for e in ind_ents if e]
```

```python
# basic similarity lookup
# gotta do something about those numbers
def find_similar_vector(doc):
    print(doc)
    max_sim = 0
    similar_ent = None
    for other in ind_ents:
        if other == doc:
            continue
        similarity = doc.similarity(other)
        if similarity > max_sim:
            max_sim = similarity
            similar_ent = other
    print(max_sim)
    print(similar_ent)
```

```python
find_similar_vector(ind_ents[30])
```
